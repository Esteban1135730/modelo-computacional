### Ejercicio 1 (resuelto)

Calcule la siguiente sumatoria:

$$
\sum_{k=1}^{n} k
$$

#### Solución matemática

Se utiliza la fórmula de la suma de los primeros \( n \) números naturales:

$$
\sum_{k=1}^{n} k = \frac{n(n+1)}{2}
$$

#### Solución en Python


```python
def suma_naturales(n):
    return n * (n + 1) // 2

print(suma_naturales(10))
```

### Ejercicio 2 (resuelto)

Calcule la siguiente sumatoria:

$$
\sum_{k=1}^{n} (2k - 1)
$$

#### Solución matemática

Esta expresión corresponde a la suma de los primeros \( n \) números impares:

$$
\sum_{k=1}^{n} (2k - 1) = n^2
$$

#### Solución en Python


```python
def suma_impares(n):
    return sum(2*k - 1 for k in range(1, n + 1))

print(suma_impares(10))
```

## III. Ejercicios de Multiplicación de Diccionarios en Python

(Se entiende por multiplicación la operación valor a valor entre diccionarios con claves comunes).

1. Dados dos diccionarios con las mismas claves y valores numéricos, construya un tercer diccionario donde cada valor sea el producto de los valores correspondientes.
2. Multiplique todos los valores de un diccionario por un escalar dado.
3. Dado un diccionario que representa dos vectores con las mismas claves, calcule su producto punto:
   
   $$
   \sum_{i} v_i \cdot w_i
   $$

4. Dado un diccionario de precios y otro de cantidades, calcule el costo total por producto y el costo total general.


```python
#1. Cree un diccionario con los nombres de cinco estudiantes como claves y sus calificaciones finales como valores. Calcule el promedio.

estudiantes = {
    "Ana": 85,
    "Bruno": 92,
    "Carmen": 78,
    "David": 88,
    "Elena": 95
}


notas = list(estudiantes.values())

promedio = sum(notas) / len(notas)
print("Calificaciones:", estudiantes)
print("Promedio:", promedio)
print(len(notas))
```

```python
#2. Dado un diccionario de productos y precios, incremente todos los precios en un 10%.
productos = {"pan": 2.0, "leche": 1.5, "huevos": 3.0, "arroz": 2.2}
print("Precios originales:", productos)

for clave in productos:
    productos[clave] = round(productos[clave] * 1.10, 2)

print("Precios +10%:", productos)
```

```python
#3. Escriba un programa que determine la clave asociada al valor máximo en un diccionario numérico.
diccionario = {"a": 10, "b": 25, "c": 15, "d": 30}

max_valor = max(diccionario.values())

clave_max = [c for c in diccionario if diccionario[c] == max_valor][0]
print("Diccionario:", diccionario)
print("Clave con valor máximo:", clave_max, "→", max_valor)

```

```python
#4. Invierta las claves y valores de un diccionario (asuma que los valores son únicos).
original = {"uno": 1, "dos": 2, "tres": 3}
invertido = {valor: clave for clave, valor in original.items()}
print("Original:", original)
print("Invertido (valor → clave):", invertido)
```

```python
#5. Dado un diccionario con palabras y su frecuencia, elimine las palabras con frecuencia menor que 3.
palabras = {"hola": 5, "mundo": 2, "python": 4, "curso": 1}
filtrado = {p: f for p, f in palabras.items() if f >= 3}
print("Original:", palabras)
print("Frecuencia >= 3:", filtrado)
```

```python
#6. Combine dos diccionarios sumando los valores de las claves comunes.
d1 = {"a": 10, "b": 20, "c": 5}
d2 = {"b": 15, "c": 10, "d": 25}
combinado = d1.copy()
for clave in d2:
    combinado[clave] = combinado.get(clave, 0) + d2[clave]
print("d1:", d1)
print("d2:", d2)
print("Combinado (suma en comunes):", combinado)
```

```python
#7. Ordene un diccionario por sus valores de forma ascendente.
d = {"c": 30, "a": 10, "b": 20}
ordenado = dict(sorted(d.items(), key=lambda x: x[1]))
print("Original:", d)
print("Ordenado por valor (asc):", ordenado)
```

```python
#8. Verifique si una clave ingresada por el usuario existe en un diccionario dado.
d = {"nombre": "Ana", "edad": 20, "ciudad": "Lima"}
clave = "edad"  # Puedes cambiar a "apellido", "nombre", "ciudad" para probar
if clave in d:
    print(f"Sí existe. Valor: {d[clave]}")
else:
    print("No existe esa clave.")
```

## III. Multiplicación de Diccionarios

```python
#1. Producto valor a valor (mismas claves)
d1 = {"a": 2, "b": 3, "c": 4}
d2 = {"a": 10, "b": 5, "c": 2}
producto = {c: d1[c] * d2[c] for c in d1 if c in d2}
print("d1:", d1)
print("d2:", d2)
print("Producto valor a valor:", producto)
```

```python
#2. Multiplique todos los valores de un diccionario por un escalar dado.
d = {"a": 1, "b": 2, "c": 3}
escalar = 5
resultado = {c: v * escalar for c, v in d.items()}
print("Diccionario:", d)
print(f"Todos × {escalar}:", resultado)
```

```python
#3. Producto punto de dos vectores (mismas claves): v·w = v₁·w₁ + v₂·w₂ + ...
v = {"x": 1, "y": 2, "z": 3}
w = {"x": 4, "y": 5, "z": 6}
producto_punto = sum(v[c] * w[c] for c in v if c in w)
print("v:", v)
print("w:", w)
print("Producto punto v·w:", producto_punto)
```

```python
#4. Costo por producto = precio × cantidad. Total = suma de todos los costos.
precios = {"pan": 2.0, "leche": 1.5, "huevos": 3.0}
cantidades = {"pan": 3, "leche": 2, "huevos": 1}
costo_por_producto = {p: precios[p] * cantidades[p] for p in precios if p in cantidades}
total = sum(costo_por_producto.values())
print("Precios:", precios)
print("Cantidades:", cantidades)
print("Costo por producto:", costo_por_producto)
print("Costo total:", total)
```
